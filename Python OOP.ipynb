```python
class Human():
    number_of_children=0
    pop_list=[]
    #Class method to increase number of individuals in Human population  
    @classmethod
    def increase_number_of_children(cls):
        cls.number_of_children=cls.number_of_children+1
    
    #Class method to append names of new individuals to population list
    @classmethod
    def add_to_list(cls):
        cls.pop_list.append(offspring)
    
    #Abstraction of human attributes
    def __init__(self,first,last,sex,blood_genotype,eye_genotype,height,weight,number_of_children,father,mother,children):
        self.first= first
        self.last = last
        self.sex  = sex
        self.blood_genotype = blood_genotype
        self.eye_genotype    = eye_genotype
        self.height     = height
        self.weight     = weight   
        self.number_of_children = number_of_children
        self.father=father
        self.mother=mother
        self.children=children
    
    def add_offspring(self,other):
        self.number_of_children =self.number_of_children+1
        other.number_of_children=other.number_of_children+1
        self.children.append(offspring)
        
    #Magic method to change instance representation
    def __repr__(self):
        return self.first

    #Magic method to spawn new offspring using '+' operator
    def __add__(self,other):
        if(self.sex=='male' and other.sex=='female'):
            return create_human(self,other)
            add_offspring(self,other)
            
        elif(self.sex=='female' and other.sex=='male'):
            return create_human(self,other)
            add_offspring(self,other)
            
        else:
            print('Both participants of the same sex. No children spawned')
            pass
```

```python
def create_genotype(x,y):
    i=np.random.randint(0,2)
    j=np.random.randint(0,2)
    x[i]+y[j]
    return(x[i]+y[j])

def blood_comb(x,y):
    if x=='O':
        return (y)
    if y=='O':
        return (x)
    if x==y:
        return (x)
    else:
        return ('AB')

def eye_comb(m,n):
    if m=='b':
        return (n)
    if n=='b':
        return (m)
    if m==n:
        return (m)
    else:
        return ('B')
    
def find_phenotype(x,y,m,n):
    blood_pheno=create_genotype(x,y)
    eye_pheno=create_genotype(m,n)
    x=blood_pheno[0]
    y=blood_pheno[1]
    m=eye_pheno[0]
    n=eye_pheno[1]
    return(blood_comb(x,y),eye_comb(m,n))
```
